# 변수 이름 정하기

print = 13을 실행하게 되면 더 이상 print() 함수를 사용할 수 없습니다.

```python
print = 13
```

```python
print('Hello World')
```

print() 함수를 다시 사용하기 위해서는 노트북 파일을 종료 후 다시 실행해야 합니다.

## 함수의 예

```python
min(4, 10, -1) # min()은 여러개의 값을 입력받은 후, 그 들중 최솟값을 반환합니다.
```

```python
max(4, 10, -1)  # max()는 여러개의 값을 입력받은 후, 그 들중 최댓값을 반환합니다.
```
