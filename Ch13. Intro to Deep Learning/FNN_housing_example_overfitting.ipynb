```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
(train_data, train_targets), (test_data, test_targets) =keras.datasets.boston_housing.load_data()
```

```python
mean = train_data.mean(axis=0)
train_data -= mean
std = train_data.std(axis=0)
train_data /= std
test_data -= mean
test_data /= std
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
train_data.shape
```

```python
model = models.Sequential()
```

```python
model.add(layers.Dense(128, activation = 'relu', kernel_initializer = 'glorot_uniform', input_shape=(train_data.shape[1],)))
```

```python
model.add(layers.Dense(1))
```

```python
model.summary()
```

```python
rmsprop = tf.keras.optimizers.RMSprop(learning_rate=0.0001, rho=0.85, momentum=0.9)
```

```python
model.compile(optimizer=rmsprop, loss='mse')
```

```python
history = model.fit(train_data, train_targets, epochs=1000, batch_size=64, validation_split=0.1)
```

## Loss   

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
```

```python
y_pred = model.predict(test_data)
from sklearn.metrics import r2_score
r2_score(test_targets, y_pred)
```

```python
y_pred = model.predict(test_data)
from sklearn.metrics import mean_squared_error
mean_squared_error(test_targets, y_pred)**0.5
```
