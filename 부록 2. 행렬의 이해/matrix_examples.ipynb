## 역행렬 구하기

```python
A = np.array([[1, 2],
              [2, 5]])
```

```python
A
```

```python
A_inv = np.linalg.inv(A)
```

```python
A_inv
```

```python
np.dot(A, A_inv)
```

```python
np.dot(A_inv,A)
```

```python
np.linalg.det(A) # determinant (행렬식)
```

## Rank 구하기

```python
A = np.array([[1, 2],
              [2, 5]])
```

```python
np.linalg.matrix_rank(A)
```

```python
B = np.array([[1, 2],
             [2, 4]])
```

```python
np.linalg.matrix_rank(B)
```

```python
np.linalg.inv(B)
```
